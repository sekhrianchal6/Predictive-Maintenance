```python
#Visualation of data correlation

sns.heatmap(data.corr(), annot = True)
plt.rcParams['figure.figsize'] = (40,20)
plt.show()
```

```python
#data grouped by engine number
grouped_data = data.groupby('Engine')
grouped_data.head()
```

```python
# Sensor state vs each number of cycles

sensor_data = data.drop(['Engine','Remaining Cycles'], axis=1)
sensor_data.set_index("Cycles").plot(subplots=True, figsize =(16,30));
```

```python
#Average of each engine performance
avg_data = data.groupby('Engine').mean().sort_values(by='Cycles', ascending = False)
avg_data_transposed = avg_data.transpose()
avg_data_transposed
```

```python
#Distribution od Remaining Cycles
plt.figure(figsize=(7, 6))
sns.histplot(data[['Remaining Cycles']]),plt.title('Remaining Cycles')
plt.show()
```
